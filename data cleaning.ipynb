```python
import pandas as pd
import matplotlib.pyplot as plt
import lazypredict
```

```python
# récupération des noms des fichiers .csv

files = []
for i in range(2001,2013):
    files.append(str(i)+'.xls')
for i in range(2013,2023):
    files.append(str(i)+'.xlsx')
    
files
```

```python
# création d'un df unique 
df = pd.DataFrame()
for i in files:
    dfi = pd.read_excel(i)
    df = pd.concat([df, dfi])
    
df
```

```python
df = df.reset_index(drop = True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.columns
```

```python
# drop useless betting odds columns 
columns_to_drop = ['CBW', 'CBL', 'GBW', 'GBL', 'IWW', 'IWL', 'SBW', 'SBL', 
                   'B&WW', 'B&WL', 'EXW', 'EXL', 'PSW', 'PSL', 'WPts',
                   'LPts', 'UBW', 'UBL', 'LBW', 'LBL', 'SJW', 'SJL',
                   'MaxW', 'MaxL','AvgW', 'AvgL']

df = df.drop(columns = columns_to_drop)
```

```python
df['Comment'].unique()
```

```python
df = df[(df['Comment'] == 'Completed') | (df['Comment'] == 'Awarded')]
df = df.drop(df[df['B365W'].isnull()].index, axis = 0)
df = df.drop(df[df['LRank'].isnull()].index, axis = 0)
df = df.drop(df[df['WRank'].isnull()].index, axis = 0)
df = df.drop(df[df['Best of'].isnull()].index, axis = 0)
df = df.drop(df[df['W2'].isnull()].index, axis = 0)
df.shape
```

```python
# converting the string to datetime format
df['Date'] = pd.to_datetime(df['Date'], format = "%Y-%m-%d")
```

```python
df['Month'] = df['Date'].dt.month
df.head()
```

```python
pd.set_option('display.max_columns', 100)
df
```

```python
df.head()
```

```python
df['FavRank'] = df['WRank'] - df['LRank']
df['FavRank'] = df['FavRank'].apply(lambda x : 'FavRank' if x < 0 else 'NoFavRank')
```

```python
df['FavRank'].value_counts()
```

```python
df['FavOdd'] = df['B365W'] - df['B365L']
df['FavOdd'] = df['FavOdd'].apply(lambda x : 'FavOdd' if x < 0 else 'NoFavOdd' if x > 0 else 'Tie')
```

```python
df['FavOdd'].value_counts()
```

```python
# encoding non numeric values to numeric values

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown="ignore", sparse=False)
ohe.fit(df[['Series']])
df[ohe.get_feature_names_out(["Series"])] = ohe.transform(df[["Series"]])
```

```python
ohe = OneHotEncoder(handle_unknown="ignore", sparse=False)
ohe.fit(df[['Court']])
df[ohe.get_feature_names_out(["Court"])] = ohe.transform(df[["Court"]])
```

```python
ohe = OneHotEncoder(handle_unknown="ignore", sparse=False)
ohe.fit(df[['Surface']])
df[ohe.get_feature_names_out(["Surface"])] = ohe.transform(df[["Surface"]])
```

```python
ohe = OneHotEncoder(handle_unknown="ignore", sparse=False)
ohe.fit(df[['Round']])
df[ohe.get_feature_names_out(["Round"])] = ohe.transform(df[["Round"]])
```

```python
df['GapOdd'] = abs(df['B365W'] - df['B365L'])
df['GapRank'] = abs(df['WRank'] - df['LRank'])
```

```python
liste = [df['GapOdd'], df['GapRank']]
for i in range(len(liste)):
    plt.subplot(1,2,i + 1)
    plt.boxplot(x = liste[i])
```

```python
liste = [df['GapOdd'], df['GapRank']]
for i in range(len(liste)):
    plt.subplot(1,2,i + 1)
    plt.hist(x = liste[i])
```

```python
from sklearn.preprocessing import RobustScaler

rs = RobustScaler()
df['GapOddScaled'] = rs.fit_transform(df[['GapOdd']])
df['GapRankScaled'] = rs.fit_transform(df[['GapRank']])
```

```python
# predicting win by rank

from sklearn.model_selection import train_test_split

X = df[(df['Best of'] == 3) & (df['GapRank'] > 50)][['Series_ATP250', 'Month', 'Series_ATP500', 'Series_Grand Slam', 'Series_International',
       'Series_International Gold', 'Series_Masters', 'Series_Masters 1000',
       'Series_Masters Cup', 'Court_Indoor', 'Court_Outdoor', 'Surface_Carpet',
       'Surface_Clay', 'Surface_Grass', 'Surface_Hard', 'Round_1st Round',
       'Round_2nd Round', 'Round_3rd Round', 'Round_4th Round',
       'Round_Quarterfinals', 'Round_Round Robin', 'Round_Semifinals',
       'Round_The Final', 'GapOddScaled', 'GapRankScaled']]
y = df[(df['Best of'] == 3) & (df['GapRank'] > 50)]['FavRank']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 41, train_size = 0.8)
```

```python
# predicting win by odd

from sklearn.model_selection import train_test_split

X = df[(df['Best of'] == 3) & (df['GapOdd'] > 0.7) & (df['GapOdd'] < 1.2)][['Series_ATP250', 'Month', 'Series_ATP500', 'Series_Grand Slam', 'Series_International',
       'Series_International Gold', 'Series_Masters', 'Series_Masters 1000',
       'Series_Masters Cup', 'Court_Indoor', 'Court_Outdoor', 'Surface_Carpet',
       'Surface_Clay', 'Surface_Grass', 'Surface_Hard', 'Round_1st Round',
       'Round_2nd Round', 'Round_3rd Round', 'Round_4th Round',
       'Round_Quarterfinals', 'Round_Round Robin', 'Round_Semifinals',
       'Round_The Final', 'GapOddScaled', 'GapRankScaled']]
y = df[(df['Best of'] == 3) & (df['GapOdd'] > 0.7) & (df['GapOdd'] < 1.2)]['FavRank']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 41, train_size = 0.8)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier

modelLR = SGDClassifier()

modelLR.fit(X_train, y_train)
print('train score :', modelLR.score(X_train, y_train))
print('test score : ', modelLR.score(X_test, y_test))
```

```python
score_test = []
abscisses = []

for i in range(15):
   X = df[(df['Best of'] == 3) & (df['GapRank'] > i * 10)][['Series_ATP250', 'Month', 'Series_ATP500', 'Series_Grand Slam', 'Series_International',
       'Series_International Gold', 'Series_Masters', 'Series_Masters 1000',
       'Series_Masters Cup', 'Court_Indoor', 'Court_Outdoor', 'Surface_Carpet',
       'Surface_Clay', 'Surface_Grass', 'Surface_Hard', 'Round_1st Round',
       'Round_2nd Round', 'Round_3rd Round', 'Round_4th Round',
       'Round_Quarterfinals', 'Round_Round Robin', 'Round_Semifinals',
       'Round_The Final', 'GapOddScaled', 'GapRankScaled']]
   y = df[(df['Best of'] == 3) & (df['GapRank'] > i * 10)]['FavRank']
    
   X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 41, train_size = 0.8)
   
   modelLR = SGDClassifier()
   modelLR.fit(X_train, y_train)
   score_test.append(modelLR.score(X_test, y_test))
   abscisses.append(i * 10)
   
plt.plot(abscisses, score_test)
plt.title('predicting FavRank with best of 3 & gap rank between 0 and 150');
```
